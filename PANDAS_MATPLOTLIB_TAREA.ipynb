**TAREA(3%): Asumiendo que cada fila en la columna del título de la PELÍCULA tiene el mismo formato:**

    Film Title Name (Year)
    
**Crea una nueva columna que pueda extraer el año de las cadenas de título y establece esta nueva columna como YEAR**

```python
# Extraer el año de los títulos de las películas
# Los títulos tienen el formato: Film Title Name (Year)
# Usamos expresiones regulares para extraer el año entre paréntesis

import re

# Función para extraer el año
def extract_year(title):
    pattern = r'\((\d{4})\)'  # Buscar 4 dígitos entre paréntesis
    match = re.search(pattern, title)
    if match:
        return int(match.group(1))
    else:
        return None

# Aplicar la función a la columna FILM
fandango['YEAR'] = fandango['FILM'].apply(extract_year)

# Verificar las primeras filas con la nueva columna
fandango[['FILM', 'YEAR']].head()
```

**TAREA(4%): Crea un gráfico KDE (o múltiples kdeplots) que muestre la distribución de las calificaciones que se muestran (STARS) versus cuál fue la calificación real de los votos (RATING). Limita los KDEs al rango 0-5.**

```python
# Crear un gráfico KDE para comparar distribuciones de STARS y RATING
plt.figure(figsize=(10, 6))

# Graficar KDE para STARS y RATING
sns.kdeplot(data=fandango_voted, x='STARS', label='STARS (Mostradas)', fill=True, alpha=0.5)
sns.kdeplot(data=fandango_voted, x='RATING', label='RATING (Real)', fill=True, alpha=0.5)

# Configuración del gráfico
plt.title('Comparación de Distribuciones: STARS vs RATING', fontsize=16)
plt.xlabel('Calificación (0-5)', fontsize=14)
plt.ylabel('Densidad', fontsize=14)
plt.xlim(0, 5)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# KDE (Kernel Density Estimation) es una forma de estimar la función de densidad de probabilidad
# de una variable aleatoria. Nos ayuda a visualizar la distribución de los datos de manera suavizada
# y continua, en lugar de usar un histograma discreto. Es especialmente útil para comparar
# distribuciones de diferentes variables o grupos.
```

**TAREA(3%): Ahora vamos a cuantificar realmente esta discrepancia. Crea una nueva columna de la diferencia entre las STARS mostradas versus el RATING verdadero. Calcula esta diferencia con STARS-RATING y redondea estas diferencias al punto decimal más cercano.**

###Normalizar columnas a STARS y RATINGS de Fandango 0-5
Observa que RT, Metacritic e IMDB no usan una puntuación entre 0-5 estrellas como lo hace Fandango. Para hacer una comparación justa, necesitamos normalizar estos valores para que todos estén entre 0-5 estrellas y la relación entre las reseñas se mantenga igual.

**TAREA(2%): Crea nuevas columnas normalizadas para todas las calificaciones para que coincidan con el rango de 0-5 estrellas mostrado en Fandango. Hay muchas formas de hacer esto.**

Enlace de ayuda: https://stackoverflow.com/questions/26414913/normalize-columns-of-pandas-data-frame

Pista más fácil:

Ten en cuenta, una forma simple de convertir calificaciones:

100/20 = 5
10/2 = 5

```python
# Normalizar las calificaciones a escala 0-5

# RT y Metacritic están en escala 0-100, dividimos por 20
combined['RT_norm'] = combined['RottenTomatoes'] / 20
combined['RT_user_norm'] = combined['RottenTomatoes_User'] / 20
combined['Metacritic_norm'] = combined['Metacritic'] / 20
combined['Metacritic_user_norm'] = combined['Metacritic_User'] / 20

# IMDB está en escala 0-10, dividimos por 2
combined['IMDB_norm'] = combined['IMDB'] / 2

# Fandango ya está en escala 0-5, pero incluimos para completitud
combined['Fandango_Stars_norm'] = combined['STARS']
combined['Fandango_Rating_norm'] = combined['RATING']

# Verificar las columnas normalizadas
norm_columns = ['FILM', 'RT_norm', 'RT_user_norm', 'Metacritic_norm', 
               'Metacritic_user_norm', 'IMDB_norm', 
               'Fandango_Stars_norm', 'Fandango_Rating_norm']

combined[norm_columns].head()
```

**TAREA(2%): Ahora crea un DataFrame norm_scores que solo contenga las calificaciones normalizadas. Incluye tanto STARS como RATING de la tabla original de Fandango.**

```python
# Crear un DataFrame con solo las calificaciones normalizadas
norm_scores = combined[['FILM', 'RT_norm', 'RT_user_norm', 'Metacritic_norm', 
                        'Metacritic_user_norm', 'IMDB_norm', 
                        'Fandango_Stars_norm', 'Fandango_Rating_norm']]

# Verificar estadísticas descriptivas de las calificaciones normalizadas
norm_scores.describe()
```

Hay muchas formas de hacer esto, pero explora la documentación de KDEplot de Seaborn para encontrar algunas formas simples de mostrarlo rápidamente. No te preocupes si el formato de tu gráfico no se ve exactamente igual al nuestro, siempre y cuando las diferencias en la distribución sean claras.**

Nota rápida si tienes problemas moviendo la leyenda para un kdeplot de seaborn: https://github.com/mwaskom/seaborn/issues/2280

```python
def move_legend(ax, new_loc, **kws):
    old_legend = ax.legend_
    handles = old_legend.legendHandles
    labels = [t.get_text() for t in old_legend.get_texts()]
    title = old_legend.get_title().get_text()
    ax.legend(handles, labels, loc=new_loc, title=title, **kws)
```

```python


# Crear un DataFrame largo para facilitar el uso de seaborn
norm_melt = pd.melt(norm_scores, id_vars=['FILM'], 
                    value_vars=['RT_norm', 'RT_user_norm', 'Metacritic_norm', 
                              'Metacritic_user_norm', 'IMDB_norm', 
                              'Fandango_Stars_norm', 'Fandango_Rating_norm'],
                    var_name='Source', value_name='Rating')

# Crear un gráfico KDE para todas las fuentes
plt.figure(figsize=(12, 8))
ax = sns.kdeplot(data=norm_melt, x='Rating', hue='Source', fill=True, alpha=0.3, 
                common_norm=False, palette='Set1')

# Añadir título y etiquetas
plt.title('Distribución de Calificaciones Normalizadas por Sitio', fontsize=18)
plt.xlabel('Calificación Normalizada (0-5)', fontsize=14)
plt.ylabel('Densidad', fontsize=14)
plt.grid(True, alpha=0.3)

# Mover la leyenda
#move_legend(ax, 'upper left')

plt.tight_layout()
plt.show()
```
